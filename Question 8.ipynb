# Finetuning GPT2 to Generate Lyrics for the prompt "I love Deep Learning"

### Load Libraries

```python
import os
import pandas as pd

%tensorflow_version 1.x

try:
  import gpt_2_simple as gpt2
except:
  !pip3 -q install gpt-2-simple
  import gpt_2_simple as gpt2
```

```python
# Fetch Lyrics (subset of data from LyricsFreak)
dfs = []
link = ('https://raw.githubusercontent.com/ThomasVrancken/'
        'lyrics_generation/master/songdata_0.csv')
for i in range(4):
    dfs.append(pd.read_csv(link.format(i)))
df = pd.concat(dfs).reset_index(drop=True)
```

```python
if not os.path.exists('content'):
    os.makedirs('content')
pd.DataFrame({"lyrics": df['text']}).to_csv(os.path.join('content', 'lyrics.csv'), index=False)
```

Using the 124M model. Could not finetune the larger models in Colab due to compute limitations

```python
gpt2.download_gpt2(model_name="124M")
```

### Hyperparameters

```python
learning_rate = 0.0001
optimizer = 'adam' # adam or sgd
batch_size = 1
model_name = "124M" # Has to match one downloaded locally
sess = gpt2.start_tf_sess()
```

### Finetune on Lyrics

```python
gpt2.finetune(sess,
              'content/lyrics.csv',
              model_name=model_name,
              sample_every=50,
              save_every=50,
              print_every=10,
              learning_rate=learning_rate,
              batch_size=batch_size,
              restore_from='latest',
              steps=500)   # max number of training steps
```

### Generate Lyrics from Prompt

There are 6 code cells below which explore different settings for the parameters such as temperature and top_p.


Reasonable lyrics are attained finally, with high `temperature` and high `top_p`.

```python
lst_results=gpt2.generate(
    sess,
    prefix="<|startoftext|>I love Deep Learning",
    nsamples=10,
    temperature=0.7,
    top_p=0.6,
    return_as_list=True,
    truncate="<|endoftext|>",
    include_prefix=True
    )

for res in lst_results:
  print(res)
  print('\n -------//------ \n')
```
